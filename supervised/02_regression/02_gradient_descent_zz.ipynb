```python
import numpy as np
import pandas as pd
import plotly.express as px

# from sklearn import
```

```python
x1 = np.array([1, 2, 3, 4, 5, 6])
y = np.array([3000, 3250, 3500, 3750, 4000, 4250])
m = len(x1)
```

```python
x1 = x1.reshape(m, 1)
y = y.reshape(m, 1)
```

```python
bias = np.ones((m, 1))
bias
```

```python
x = np.append(bias, x1, axis=1)
x
```

## Losowa initializacja parametrów.



```python
eta = 0.01 # Wskaźnik uczenia
weights = np.random.randn(2, 1)
weights
```

## Metoda gradientu prostego

```python
# y = a*x + b
intercept = [] # b
coef = [] # a

for i in range(3000):
  gradient = (2/m) * x.T.dot(x.dot(weights) - y)
  weights = weights - eta * gradient
  intercept.append(weights[0][0])
  coef.append(weights[1][0])

print(weights)
```

```python
df = pd.DataFrame(data={"intercept": intercept, "coef": coef})
df.head()
```

## Wizualizacja dopasowania

```python
px.line(df, y="intercept", width=800, title="Fit: intercept")
```

```python
px.line(df, y="coef", width=800, title="Fit: coef")
```
